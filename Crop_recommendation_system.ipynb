```python
import numpy as np
import pandas as pd
from sklearn.linear_model import SGDClassifier
```

```python
dataframe = pd.read_csv('Crop_recommendation.csv')
```

```python
dataframe = dataframe.sample(frac=1, random_state=42).reset_index(drop=True)
print(dataframe)
```

```python
params = dataframe.drop(columns=['label'])
```

```python
print(params)
```

```python
crop_recommended = dataframe['label']
crop_recommended
```

```python
params_train, params_test, crop_recommended_train, crop_recommended_test= params[:1850], params[1850:], crop_recommended[:1850],crop_recommended[1850:]
```

```python
sgd_clf  = SGDClassifier(random_state=42)
sgd_clf.fit(params_train, crop_recommended_train)
```

```python
sgd_clf.predict([params_test.iloc[0]])
```
